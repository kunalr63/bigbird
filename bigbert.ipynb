```python
col1=[]#sentence id
col2=[]#word
col3=[]#tags
for i in range(len(data)):
  txt=data['sentences'][i]
  tag=data['tags'][i]
  for j in range(len(txt)):
    col1.append(i)
    col2.append(txt[j])
    col3.append(tag[j])
```

```python
dic={
    "sentence_id":col1,
     "words":col2,
     "labels":col3
}
```

```python
my_data= pd.DataFrame(dic)
my_data
```

```python
X= my_data[["sentence_id","words"]]
Y =my_data["labels"]
```

```python
train_data=my_data[:60000]
test_data=my_data[60000:]
```

# Model Training


```python
from simpletransformers.ner import NERModel,NERArgs
```

```python
label = my_data["labels"].unique().tolist()
label
```

```python
# BigBird-base-cased learning rate 

args = NERArgs()
# args.wandb_project='new_projects'
args.num_train_epochs = 20
args.learning_rate =5e-5
args.overwrite_output_dir =True
args.save_model_every_epoch=False
args.max_seq_length=4096
args.train_batch_size = 1
args.eval_batch_size = 0
```

```python
model3 = NERModel('bigbird', 'google/bigbird-roberta-base',labels=label,args =args,ignore_mismatched_sizes=True)
```

```python
# model3 = NERModel('xlnet', 'xlnet-large-cased',labels=label,args =args,ignore_mismatched_sizes=True,use_cuda=True)
import torch
torch.cuda.empty_cache()
```

```python
model3.train_model(train_data,eval_data = test_data,acc=accuracy_score)  
```
